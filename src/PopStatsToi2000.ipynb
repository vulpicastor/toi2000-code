```python
def calculate_surface_grav(grav_param, radius):
    return (grav_param / radius ** 2).cgs

def calculate_density(mass, radius):
    density = mass / (4/3 * np.pi * radius ** 3)
    return density.cgs

def calculate_density_err(density, mass, mass_err, radius, radius_err):
    density_var = ((mass_err / mass) ** 2
                   + 3 * (radius_err / radius) ** 2) * density ** 2
    return np.sqrt(density_var)

# def calculate_time_ecc_damp(mass_planet, radius_planet, mass_star, period, qual_factor=1e5):
#     tau_ep = (5. * qual_factor * (mass_planet / const.M_jup)
#               * (mass_star / const.M_sun) ** (2 / 3.)
#               * (period / (1. * u.day)) ** (13 / 3.)
#               * (radius_planet / const.R_jup) ** -5) * u.year
#     return (tau_ep.si).to(u.year)

def calculate_time_ecc_damp(planet_grav_param, planet_radius, star_grav_param, period, qual_factor=1e5):
    tau_ep = ((4. / 63) * qual_factor * planet_grav_param
              * star_grav_param ** (2. / 3)
              * planet_radius ** (-5)
              * (period / 2 / np.pi) ** (13. / 3))
    return tau_ep.to(u.year)
```
